```python
import pandas as pd
import numpy as np
from google.colab import files

uploaded = files.upload()

# Main dataset
opportunities = pd.read_csv('data_cleaned_final.csv')

```

```python
# candidateTags dataset

uploaded = files.upload()
tags_df = pd.read_csv('CandidateTags.csv')
tags_list = tags_df.iloc[:, 0].tolist()
```

```python
from sentence_transformers import SentenceTransformer, util
import torch

# Load the model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Convert descriptions and tags to embeddings
description_embeddings = model.encode(opportunities['Description'].tolist(), convert_to_tensor=True)
tag_embeddings = model.encode(tags_list, convert_to_tensor=True)

# Compute cosine similarities
cosine_scores = util.pytorch_cos_sim(description_embeddings, tag_embeddings)

# Find the best matching tag for each description
best_matching_tags = [tags_list[torch.argmax(cs).item()] for cs in cosine_scores]

# Add the best matching tags to the opportunities data
opportunities['BestMatchingTag'] = best_matching_tags


```

```python
# Randomly sample 10 entries from the opportunities dataset
sample_reviews = opportunities.sample(10)
print(sample_reviews[['Description', 'BestMatchingTag']])
```
